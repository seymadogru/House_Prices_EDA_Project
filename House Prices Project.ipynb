# Tablodan bir sütun seçip Outliers örneği yapalım.

```python
df = pd.read_csv('house_prices_train.csv')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
title_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
axis_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.boxplot(df.SalePrice.dropna())
plt.title("Sales Prices (whis=1.5)", fontdict=title_font)
plt.show()
```

```python
q75, q25 = np.percentile(df.SalePrice.dropna(), [75 ,25])
IQR = q75 - q25
threshold_values=[]
for threshold_value in np.arange(1,5,0.5):
    min_value = q25 - (IQR*threshold_value)
    max_value = q75 + (IQR*threshold_value)
    outlier_value_number = len((np.where((df.SalePrice.dropna() > max_value)| (df.SalePrice.dropna() < min_value))[0]))
    threshold_values.append({'threshold_value': threshold_value,'outlier_value_number': outlier_value_number})
pd.DataFrame(threshold_values)
```

```python
plt.hist(df.SalePrice)
plt.title("Sales Prices", fontdict=title_font)
plt.show()
```

Şimdi aykırı değerleri silerek başlayalım. Yukarıdaki threshold tablosuna baktığımda ben threshold value yu 1 almaya karar verdim.

```python
def outliers(df, ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1 * IQR
    upper_bound = Q3 + 1 * IQR
    ls = df.index[(df[ft] < lower_bound) | (df[ft]>upper_bound)]
    return ls
```

```python
index_list = []
for feature in ['SalePrice']:
    index_list.extend(outliers(df , feature))
print(len(index_list))
print(index_list)
```

```python
def remove(df , ls):
    ls = sorted(set(ls))
    df=df.drop(ls)
    return df
```

```python
df1 = remove(df , index_list)
```

### Normalization ve Standardization uygulayalım.

```python
from sklearn.preprocessing import normalize

df["SalePrice_Normalize"] = normalize(np.array(df["SalePrice"]).reshape(1,-1)).reshape(-1,1)

normal_features=["SalePrice_Normalize","SalePrice"]

print('Minimum Value;\n',)
print(df[normal_features].min())
print("-------------------------------------")
print('\nMaximum Value;\n',)
print(df[normal_features].max())
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df[normal_features[0]], df[normal_features[1]])
plt.ylabel("Sale Price Value", fontdict=axis_font)
plt.xlabel("Normalized Sale Price Value", fontdict=axis_font)
plt.show()
```

```python
from sklearn.preprocessing import scale

df["scaled_SalePrice"] = scale(df["SalePrice"])
scaled_features=["scaled_SalePrice","SalePrice"]

print('Standard Deviations\n-----------------',)
print(df[scaled_features].std())
print('\nOrtalamalar\n-----------------',)
print(df[scaled_features].mean())
```

```python
plt.scatter(df[scaled_features[0]], df[scaled_features[1]])
plt.ylabel("Original Value", fontdict=axis_font)
plt.xlabel("Standardized  Value", fontdict=axis_font)

plt.show()
```
